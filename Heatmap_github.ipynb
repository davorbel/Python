```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt

# Use seaborn style
sns.set()

# Read CSV from GitHub raw URL
url = 'https://github.com/davorbel/Python/raw/main/P%20den_proteome%20data.xlsx'
df = pd.read_excel(url)

print(df.keys())
print(df.shape)
df


df.dropna(inplace = True)
print(df.shape)
df = df.set_index('Protein_ID')

sns.clustermap(df, metric="euclidean", z_score=0, method="average", cmap="cividis")
plt.savefig("Heat_Pden1.pdf")
plt.show()


# Change color palette
#sns.clustermap(df, metric="euclidean", z_score=0, method="average", cmap="mako")
#plt.show()
#sns.clustermap(df, metric="euclidean", z_score=0, method="average", cmap="viridis")
#plt.show()
#sns.clustermap(df, metric="euclidean", z_score=0, method="average", cmap="Blues")
#plt.show()
```

Seaborn offers a wide range of colormaps. Some popular perceptual colormaps include `mako`, `rocket`, `viridis`, `plasma`, `magma`, and `cividis`. You can also use sequential colormaps like `Blues`, `Greens`, `Reds`, or diverging colormaps like `coolwarm`, `bwr`, `seismic`.
